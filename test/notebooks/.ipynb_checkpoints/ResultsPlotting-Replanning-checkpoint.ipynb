# Display experimental results

```julia
function construct_replanner_result_dataframe(problem_dir,results_dir,N_problems)
    config_df = construct_config_dataframe(problem_dir,N_problems,:replanning_config,(dir,id)->joinpath(dir,string("stream",id,"/stream_config.toml")))
    modes = [:replanning]
    Dict(
        mode => join(config_df,
            construct_result_dataframe(mode, problem_dir,results_dir,N_problems,(dir,mode,id)->joinpath(dir,string("stream",id,"/results.toml"))),
            on = :problem_id, kind = :inner) for mode in modes
    )
end
```

```julia
function plot_histories(df;
        y_key=:makespans,
        m_key=:M,
        m_vals=sort(unique(df[m_key])),
        c_key=:arrival_interval,
        c_vals=sort(unique(df[c_key])),
        opt_key=:optimal,
        size=2pt,
        shape=Shape.xcross,
        geom1=Geom.line,
        geom2=Geom.point,
        geom3=Geom.point,
        scale=Scale.y_log10(format=:plain),
        ymin=minimum(df[y_key]),
        ymax=maximum(df[y_key]),
        color_scale = Scale.color_discrete_hue(),
        colors_vec = color_scale.f(length(c_vals)),
        colors_dict = Dict(c=>colors_vec[i] for (i,c) in enumerate(c_vals))
    )
    layers = []
    for (i,m_val) in enumerate(m_vals)
        df_cut = df[df[m_key] .== m_val,:]
        for j in 1:nrow(df_cut)
            arr = df_cut[y_key][j]
            color = colors_dict[df_cut[c_key][j]]
            opt_vals = df_cut[opt_key][j]
            idx = 1
            while idx < length(arr)
                if arr[idx+1] > 0
                    idx += 1
                else
                    break
                end
            end
            # Overall trend
            push!(layers,layer(
                    x=1:idx,
                    y=arr[1:idx],
                    geom1,
                    Theme(default_color=color)
                    ))
            # Timeouts
            tmidxs=[i for i in 1:idx if ~opt_vals[i]]
            push!(layers,layer(
                    x=tmidxs, 
                    y=arr[tmidxs],
                    shape=[shape],
                    size=[size],
                    Geom.point,
                    Theme(default_color=color)
                    ))
            if idx < length(arr)
                # End points
                push!(layers,layer(
                        x=[idx],
                        y=[arr[idx]],
                        geom2,
                        Theme(default_color=colors_vec[i])
                        ))
            end
        end
    end
#     plot(layers...,Coord.cartesian(ymin=ymin, ymax=ymax))
    plot(layers...,scale,Coord.cartesian(ymin=ymin, ymax=ymax))
end
```

```julia
set_default_plot_size(28cm,12cm)
N_problems = 36
# base_dir            = joinpath(EXPERIMENT_DIR,"replanning/long_problems")
base_dir            = joinpath(EXPERIMENT_DIR,"replanning")
base_problem_dir    = joinpath(base_dir,"problem_instances")
base_results_dir    = joinpath(base_dir,"results")


replan_model    = MergeAndBalance()
# replan_model    = Oracle()
# replan_model    = ReassignFreeRobots()
# replan_model    = DeferUntilCompletion()
fallback_model  = ReassignFreeRobots()
problem_dir     = base_problem_dir
results_dir     = joinpath(base_results_dir, string(typeof(replan_model),"-",typeof(fallback_model)))

df_dict = construct_replanner_result_dataframe(problem_dir,results_dir,N_problems);
```

```julia
set_default_plot_size(28cm,8cm)

df = df_dict[:replanning]
ymin=1.35
ymax=3
hstack(
    plot_histories(df[(df.M .== 10),:];ymin=ymin,ymax=ymax),
    plot_histories(df[(df.M .== 20),:];ymin=ymin,ymax=ymax),
    plot_histories(df[(df.M .== 30),:];ymin=ymin,ymax=ymax),
)
```
